```python
# ============================================
#           LIBRARY MANAGEMENT SYSTEM
# ============================================

# ---- Data Storage ----
library  = {}   # all books
borrowed = {}   # borrowed books
members  = {}   # registered members

# ============================================
#              BOOK FUNCTIONS
# ============================================

def add_book(title, author, year, copies):
    if title in library:
        library[title]['copies'] += copies
        print(f"'{title}' stock updated! Total copies: {library[title]['copies']}")
    else:
        library[title] = {
            'author' : author,
            'year'   : year,
            'copies' : copies,
            'info'   : (title, author, year)   # tuple — fixed info
        }
        print(f"'{title}' added successfully!")

def remove_book(title):
    if title in library:
        del library[title]
        print(f"'{title}' removed from library!")
    else:
        print(f"'{title}' not found!")

def show_all_books():
    if not library:
        print("No books in library!")
        return
    print("\n========== ALL BOOKS ==========")
    for title, details in library.items():
        status = "Available" if details['copies'] > 0 else "Not Available"
        print(f"Title   : {title}")
        print(f"Author  : {details['author']}")
        print(f"Year    : {details['year']}")
        print(f"Copies  : {details['copies']}")
        print(f"Status  : {status}")
        print("--------------------------------")

# ============================================
#             MEMBER FUNCTIONS
# ============================================

def add_member(member_id, name, age):
    if member_id in members:
        print(f"Member ID {member_id} already exists!")
    else:
        members[member_id] = {
            'name'        : name,
            'age'         : age,
            'books_taken' : [],    # list of borrowed books
            'info'        : (member_id, name, age)  # tuple
        }
        print(f"Member '{name}' registered successfully!")

def show_all_members():
    if not members:
        print("No members registered!")
        return
    print("\n========== ALL MEMBERS ==========")
    for mid, details in members.items():
        print(f"ID      : {mid}")
        print(f"Name    : {details['name']}")
        print(f"Age     : {details['age']}")
        print(f"Books   : {details['books_taken']}")
        print("----------------------------------")

# ============================================
#           BORROW & RETURN FUNCTIONS
# ============================================

def borrow_book(member_id, title):
    # check member exists
    if member_id not in members:
        print(f"Member ID {member_id} not found!")
        return

    # check book exists
    if title not in library:
        print(f"'{title}' not found in library!")
        return

    # check copies available
    if library[title]['copies'] == 0:
        print(f"Sorry! '{title}' is not available right now!")
        return

    # check member already has this book
    if title in members[member_id]['books_taken']:
        print(f"{members[member_id]['name']} already has '{title}'!")
        return

    # borrow the book
    members[member_id]['books_taken'].append(title)
    library[title]['copies'] -= 1
    borrowed[f"{member_id}_{title}"] = {
        'member' : members[member_id]['name'],
        'title'  : title,
        'status' : 'borrowed'
    }
    print(f"{members[member_id]['name']} borrowed '{title}' successfully!")
    print(f"Remaining copies of '{title}': {library[title]['copies']}")

def return_book(member_id, title):
    # check member exists
    if member_id not in members:
        print(f"Member ID {member_id} not found!")
        return

    # check member actually borrowed this book
    if title not in members[member_id]['books_taken']:
        print(f"{members[member_id]['name']} never borrowed '{title}'!")
        return

    # return the book
    members[member_id]['books_taken'].remove(title)
    library[title]['copies'] += 1
    del borrowed[f"{member_id}_{title}"]
    print(f"{members[member_id]['name']} returned '{title}' successfully!")
    print(f"Copies of '{title}' now: {library[title]['copies']}")

# ============================================
#              SEARCH FUNCTIONS
# ============================================

def search_book(title):
    if title in library:
        details = library[title]
        print(f"\n--- Book Found ---")
        print(f"Title  : {title}")
        print(f"Author : {details['author']}")
        print(f"Year   : {details['year']}")
        print(f"Copies : {details['copies']}")
    else:
        print(f"'{title}' not found!")

def show_available_books():
    available = {t for t, d in library.items() if d['copies'] > 0}  # set
    print("\n========== AVAILABLE BOOKS ==========")
    if not available:
        print("No books available right now!")
        return
    for title in available:
        info = library[title]['info']
        print(f"{info[0]} by {info[1]} ({info[2]}) — {library[title]['copies']} copies")

def show_borrowed_books():
    print("\n========== BORROWED BOOKS ==========")
    if not borrowed:
        print("No books currently borrowed!")
        return
    for key, details in borrowed.items():
        print(f"{details['member']} → '{details['title']}'")

def member_books(member_id):
    if member_id not in members:
        print("Member not found!")
        return
    name  = members[member_id]['name']
    books = members[member_id]['books_taken']
    print(f"\n{name}'s borrowed books:")
    if not books:
        print("No books borrowed!")
    else:
        for book in books:
            print(f"  → {book}")

# ============================================
#                 MAIN PROGRAM
# ============================================

# adding books
add_book("Python Basics",    "Guido",       2020, 3)
add_book("Clean Code",       "Martin",      2008, 2)
add_book("Data Structures",  "Cormen",      2009, 2)
add_book("Deep Learning",    "Goodfellow",  2016, 1)
add_book("Python Basics",    "Guido",       2020, 2)   # adding more copies

# adding members
add_member(101, "Mohammad", 20)
add_member(102, "Ali",      22)
add_member(103, "Sara",     21)

# show everything
show_all_books()
show_all_members()

# borrowing books
borrow_book(101, "Python Basics")
borrow_book(101, "Clean Code")
borrow_book(102, "Deep Learning")
borrow_book(103, "Python Basics")
borrow_book(102, "Deep Learning")   # already borrowed!
borrow_book(104, "Clean Code")      # member not found!

# show borrowed
show_borrowed_books()

# show available
show_available_books()

# member wise books
member_books(101)
member_books(102)

# returning books
return_book(101, "Python Basics")
return_book(102, "Deep Learning")

# search
search_book("Clean Code")
search_book("Harry Potter")   # not found

# final available books
show_available_books()
```
